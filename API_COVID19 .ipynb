# infos de chaque pays

```python
dict_data_covid19=dict_data_covid19['Countries']
```

```python
#créer un liste vide pour stocker les données
Data_info =[]
import pandas as pd
for t in dict_data_covid19:
    Data_info.append( [t['Country'],t['CountryCode'],t['Slug'],t['NewConfirmed'],t['TotalConfirmed'],t['NewRecovered'],t['Date']])
    
```

```python
Data_info
```

```python
import pandas as pd
df1 = pd.DataFrame (data= Data_info, columns = ['Country','CountryCode','Slug','NewConfirmed','TotalConfirmed','NewRecovered','Date'])
```
